# Line Graph

```python
x = np.linspace(0,10)
y= 5*x*x*x
plt.plot(x, y)
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0,20)
y= x**2
plt.plot(x, y)
plt.xlabel('x-values')
plt.ylabel('x^2-values')
plt.title('line graph')
plt.grid(True)
#plt.show()
plt.savefig("hgfhgf.pdf", format="pdf", bbox_inches="tight")
```

```python
x = np.linspace(-10,10)
y= x**2
z= x**3
plt.plot(x,z,color='red')
plt.plot(x, y, '*')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('line plot')
plt.grid(True)
plt.show()
```

```python
plt.plot(np.sin(x),label='sin(x)',color='orange')
plt.plot(np.cos(x),label='cos(x)',color='green')
plt.xlim(10,50)
plt.legend()
plt.title('math functions')
plt.show()
```

# Sub plot

```python
fig, axs = plt.subplots(3, 3,gridspec_kw={'hspace': 0.5, 'wspace': 0.4})
x = np.linspace(0, 20, 400)
y = np.sin(x)
z=np.cos(x)
m = (x**3)
n=(x**2)
k=(4*x)
axs[0, 0].plot(x, y)
axs[0, 0].set_title('sin(x)')
axs[0, 1].plot(x, z, 'tab:orange')
axs[0, 1].set_title('cos(x)')
axs[1, 0].plot(x, m, 'tab:green')
axs[1, 0].set_title('x**3')
axs[1, 1].plot(x,n, 'tab:red')
axs[1, 1].set_title('x**2')
axs[2, 1].plot(x, k)
axs[2, 1].set_title('4X')


plt.show()
```

```python
fig, axs = plt.subplots(2, 2 ,gridspec_kw={'hspace': 0.5, 'wspace': 0.4})
x = np.linspace(0, 20, 400)
y = np.sin(x)
z=np.cos(x)
m = (x**3)
n=(x**2)
axs[0, 0].plot(x, y)
axs[0, 0].set_title('sin(x)')
axs[0, 1].plot(x, z, 'orange')
axs[0, 1].set_title('cos(x)')
axs[1, 0].plot(x, m, 'green')
axs[1, 0].set_title('x**3')
axs[1, 1].plot(x,n, 'red')
axs[1, 1].set_title('x**2')
plt.show()
```

# Pie charts

```python
a= iris['variety'].value_counts()
species = a.index
count = a.values
colors= ['lightblue','lightgreen','gold']
explode = (0,0,.6)
plt.pie(count, labels=species,shadow=True,
        colors=colors,explode = explode, autopct='%1.1f%%')
plt.xlabel('species')
plt.axis('equal')
plt.show()
```

# Seaborn(Visualization Lybrary)

```python
import seaborn as sns
```

# Line Graph

```python
sns.set_style('darkgrid')
sns.lineplot(data=iris.drop(['variety'], axis=1))
plt.show()
```

# Scatter Plot

```python
sns.FacetGrid(iris, hue="variety", height=6).map(plt.scatter, "sepal.length", "sepal.width").add_legend()
plt.show()
```

# Bar Plot

```python
a= iris['variety'].value_counts()
species = a.index
count = a.values
sns.barplot(x = species, y=count)
plt.show()
```

```python
sns.FacetGrid(iris, hue="variety", height=5) \
   .map(sns.distplot, "petal.length") \
   .add_legend()
plt.show()
```
