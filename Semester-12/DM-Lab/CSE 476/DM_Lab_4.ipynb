# Handling Missing Values

```python
data_1=data.copy()
data_1.isnull().sum()
```

# 1. Eliminating Missing Value (Row)

```python
data_without_missing_values=data_1.dropna(subset=["engV","drive"])
data_without_missing_values
```

```python
data_without_missing_values=data_without_missing_values.copy()
data_without_missing_values.isnull().sum()
```

# 2. Eliminate columns (Any column having more than 80-90% null value)

# 3.Estimate missing values

```python
from sklearn.impute import SimpleImputer
```

```python
data_1=data_1.iloc[:,:].values
```

```python
data_1
```

```python
import numpy as np
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
```

```python
imputer=imputer.fit(data_1[0:,0:10]) #1st portion is for row (start:end): 2nd portion is for columns (start:end)
```

```python
data_1[0:,0:10]=imputer.transform(data_1[0:,0:10])
data_1[0:,0:10]
data_1
```

```python
left = pd.DataFrame({"key": ["K0", "K1", "K2", "K3"],
            "A": ["A0", "A1", "A2", "A3"],
            "B": ["B0", "B1", "B2", "B3"],}) 

right = pd.DataFrame({"key": ["K0", "K1", "K2", "K3"],
            "C": ["C0", "C1", "C2", "C3"],
            "D": ["D0", "D1", "D2", "D3"],})
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, on="key")
```

```python
left = pd.DataFrame({"key1": ["K0", "K0", "K1", "K2"], "key2": ["K0", "K1", "K0", "K1"],
            "A": ["A0", "A1", "A2", "A3"],
            "B": ["B0", "B1", "B2", "B3"],}) 

right = pd.DataFrame({"key1": ["K0", "K1", "K1", "K2"], "key2": ["K0", "K0", "K0", "K0"],
            "C": ["C0", "C1", "C2", "C3"],
            "D": ["D0", "D1", "D2", "D3"],})
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, on=["key1", "key2"])
```

# Left Outer Join

```python
pd.merge(left, right, how="left", on=["key1", "key2"])
```

# Right Outer Join

```python
pd.merge(left, right, how="right", on=["key1", "key2"])
```

# Full Outer Join

```python
pd.merge(left, right, how="outer", on=["key1", "key2"])
```

# Inner Join

```python
pd.merge(left, right, how="inner", on=["key1", "key2"])
```
